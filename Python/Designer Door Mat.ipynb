### Code :

```python
N, M = (int(value) for value in input("Mat size must be NxM format. (N is an odd natural number, and M is 3 times N.)\nEnter the mat size in accordance with the above rule, leaving a space between. (Example : 9 27)\n\nPlease enter the mat size : ").split())

# Top Side and Center
for i in range(1, int((N+1)/2 + 1)) :
    if i == int((N+1)/2) :
        print("WELCOME".center(M, "-"))
    else :
        print((".|." * (i * 2 -1)).center(M, "-"))
        
# Bottom Side        
for i in range(int((N-1)/2), 0, -1) :
    print((".|." * (i * 2 -1)).center(M, "-"))
```
