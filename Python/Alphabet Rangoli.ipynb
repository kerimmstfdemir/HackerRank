### Code :

```python
n = int(input('Please enter an integer number between "0 < n < 27" : '))
alphabet = "abcdefghijklmnopqrstuvwxyz"
n_letters = alphabet[:n]
line = []

for i in range(1, len(n_letters) + 1) :
    line.append(n_letters[-i])
    
for i in range(0, len(line)) :
    print(("-".join(line[0:i+1])).rjust(len(line) * 2 - 1, "-"), end="")
    if i != 0 :
        print("-" + ("-".join(line[i-1::-1])).ljust(len(line) * 2 - 3, "-"))
    else:
        print("-" * (len(line) * 2 - 2))

for i in range(len(line)-2, -1, -1) :
    print(("-".join(line[0:i+1])).rjust(len(line) * 2 - 1, "-"), end="")
    if i != 0 :
        print("-" + ("-".join(line[i-1::-1])).ljust(len(line) * 2 - 3, "-"))
    else:
        print("-" * (len(line) * 2 - 2))
```
